```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# first 10 lines of the full csv for ease of working
file_path_small = 'data/CRIME/Crimes_-_2019_to_Present.csv'
```

```python
# parse_dates[] specificies cols which contain date information, index_col[] specifices cols to be indexed
crimes = pd.read_csv(file_path_small, parse_dates=[2], index_col=[2])
```

```python
# select primary type cols, group by day, count them
daily_crimes = crimes['Primary Type'].resample('D').count()
daily_crimes.sort_index(inplace=True)
daily_crimes.plot()
```

```python
mi = crimes.set_index(['Primary Type'], append=True)
mi
```

```python
mi.ID
```

```python
# populate mi with occurences field and fill with 1s
mi['occurences'] = np.ones(len(mi), dtype=int)
# sums each count of each primary type and outputs it
mi_group = mi.occurences.groupby(level=[0,1]).sum()
mi_group
```

```python
# separate out 1st index of MultiIndex from mi_group, fill missing with 0, resample by hour sumed, fill missing with zero, sorted by type as int (???)
crimes_by_type = mi_group.unstack(level=1).fillna(0).resample('h').sum().fillna(0).astype(int)
crimes_by_type.head()
```

```python
crimes_by_type.loc[:,['THEFT', 'HOMICIDE', 'ASSAULT']]['2019'].resample('D').sum().plot()
```
